```python
import librosa
import matplotlib.pyplot as plt
import librosa.display

# Load the audio file
filename = r"C:\Users\Tarun\Documents\Sound Recordings\Recording (2).wav"
y, sr = librosa.load(filename)

# Define start and end times for different segments
start_time_1 = 0.25
end_time_1 = 1.5
segment_1 = y[int(start_time_1 * sr):int(end_time_1 * sr)]

start_time_2 = 1.5
end_time_2 = 3
segment_2 = y[int(start_time_2 * sr):int(end_time_2 * sr)]

# Plot and display the first segment for A4
plt.figure(figsize=(12, 4))
librosa.display.waveshow(segment_1, sr=sr)
plt.title("Segment 1")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()

# Plot and display the second segment for A4
plt.figure(figsize=(12, 4))
librosa.display.waveshow(segment_2, sr=sr)
plt.title("Segment 2")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()
```
