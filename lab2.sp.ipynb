```python
import numpy as np
from scipy.io import wavfile
import matplotlib.pyplot as plt

# A1
def compute_first_derivative(signal):
    return np.diff(signal)

fs, speech_signal = wavfile.read(r"C:\Users\Tarun\Documents\Sound Recordings\Recording (2).wav")

first_derivative_signal = compute_first_derivative(speech_signal)

# A2
zero_crossings = np.where(np.diff(np.sign(first_derivative_signal)))[0]
average_length = np.mean(np.diff(zero_crossings))

time = np.arange(0, len(speech_signal)) / fs

plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(time, speech_signal, label='Original Speech Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(time[1:], first_derivative_signal, label='First Derivative')
plt.scatter(time[zero_crossings], first_derivative_signal[zero_crossings], color='red', label='Zero Crossings')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()

plt.show()

print(f"Average length between two consecutive zero crossings: {average_length} samples")
```
